```python
import speech_recognition as sr
from gtts import gTTS
import os
from datetime import datetime
import playsound
import pyjokes
import wikipedia
import pyaudio
import webbrowser
import winshell
from pygame import mixer
import pathlib
```

# Converted speech to text

```python
def speak(text):
    tts = gTTS(text=text, lang='en')
    filename = "voice.mp3"
    try:
        os.remove(filename)
    except OSError:
        pass
    tts.save(filename)
    playsound.playsound(filename)
```

# Load audio from mic

```python
#get mic audio
def get_audio():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.pause_threshold = 1
        # wait for a second to let the recognizer adjust the
        # energy threshold based on the surrounding noise level
        r.adjust_for_ambient_noise(source, duration=1)
        audio = r.listen(source)
        said = ""
        try:
            said = r.recognize_google(audio)
            print(said)
        except sr.UnknownValueError:
            speak("Sorry, I did not get that.")
        except sr.RequestError:
            speak("Sorry, the service is not available")
    return said.lower()
```

# Map keywords to actions

```python
#function to respond to commands
def respond(text):
    print("Text from get audio " + text)
    if 'youtube' in text:
        speak("What do you want to search for?")
        keyword = get_audio()
        if keyword!= '':
            url = f"https://www.youtube.com/results?search_query={keyword}"
            webbrowser.get().open(url)
            speak(f"Here is what I have found for {keyword} on youtube")

    elif 'search' in text:
        speak("What do you want to search for?")
        query = get_audio()
        if query !='':
            result = wikipedia.summary(query, sentences=3)
            speak("According to wikipedia")
            print(result)
            speak(result)

    elif 'joke' in text:
        speak(pyjokes.get_joke())

    elif 'empty recycle bin' in text:
        winshell.recycle_bin().empty(confirm=False, show_progress=False, sound=True)
        speak("Recycle bin emptied")

    elif 'what time' in text:
        strTime = datetime.today().strftime("%H:%M %p")
        print(strTime)
        speak(strTime)

    elif 'play music' in text or 'play song' in text:
        playmusic()

    elif 'stop music' in text:
        stopmusic()

    elif 'exit' in text:
        speak("Goodbye, till next time")

#play music
def playmusic():
    speak("Now playing...")
    #add your music on folder music
    music_dir = pathlib.Path.cwd() / "music"
    songs = os.listdir(music_dir)
    #play first music
    song = music_dir / songs[0]

    mixer.init()
    mixer.music.load(song)
    mixer.music.play()
    
#stop music
def stopmusic():
    speak("Stopping playback.")
    mixer.music.stop()
```

# Main code

```python
# key words:
# 'youtube'
# 'search'
# 'joke'
# 'empty recycle bin'
# 'what time'
# 'play music'
# 'stop music'
# 'exit'

# loop until say exit
text = ''
while text != 'exit':
    speak("Say a key word")
    print("I am listening...")
    text = get_audio()
    respond(text)
```
